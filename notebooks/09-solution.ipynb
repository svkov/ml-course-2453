
## Задача

Применить `RandomForestRegressor` к датасету с погодой. Подобрать гиперпараметры. Применить кросс-валидацию, сравнить с ARIMA.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
x_train, x_test = df[:pivot], df[pivot:]
t_train, t_test = np.arange(len(x_train)).reshape(-1, 1), np.arange(len(x_train), len(x_train) + len(x_test)).reshape(-1, 1)
params = {
    'n_estimators': [50, 100, 150, 500],
    'max_depth': [3, 4, 5]
}
search = RandomizedSearchCV(RandomForestRegressor(), params, scoring='r2')
search.fit(t_train, x_train)
search.best_estimator_
```

Результат получился плохим по следующим причинам:

1. Случайный лес перемешивает данные внутри себя (это не допустимо для временных рядов)
2. RandomizedSearchCV реализует кросс-валидацию, которая не адаптирована под временные ряды

```python
search.best_estimator_.score(t_test, x_test)
```

Как правильно пользоваться случайным лесом

```python
# Конвертируем параметры в удобный вид
params_to_select = [dict(zip(params.keys(), values)) for values in product(*params.values())]
params_to_select
```

```python
best_metric = 0
best_model = None
best_params = None
best_pred = None
for param in params_to_select:
    test = []
    pred = []
    for i in range(50, len(df_monthly)):
        x_train = df_monthly.t.iloc[:i].values.reshape(-1)
        test.append(df_monthly.t.iloc[i])
        t_train = np.arange(i).reshape(-1, 1)
        t_test = np.array([[i]])
        
        model = RandomForestRegressor(bootstrap=False, **param)
        model.fit(t_train, x_train)
        pred.append(model.predict(t_test))
    metric = r2_score(test, pred)
    if metric > best_metric:
        best_metric = metric
        best_params = param
        best_model = model
        best_pred = pred
```

```python
plt.plot(df_monthly.t.iloc[50:].values)
plt.plot(best_pred)
plt.legend(['test', 'pred'])
```

Попробуем спрогнозировать в будущее, захватив кусок из тестовой выборки (на которой модель обучалась)

```python
test = np.arange(90, 120).reshape(-1, 1)
plt.plot(df_monthly.t.iloc[90:].values)
plt.plot(best_model.predict(test))
plt.legend(['test', 'pred'])
```

Как видим, модель переобучилась. По неизвестным новым входным данным модель выдает просто какое-то среднее значение.

Модель случайного леса плохо подходит для прогнозирования одномерных временных рядов. Если бы были какие-то объясняющие признаки, то модель бы увидела закономерность и на их основе построила какой-то прогноз.

## Задача

Применить функцию `fourierExtrapolation` к исходному ряду `df['t']`

```python
from numpy import fft
    
def fourierExtrapolation(x, n_predict):
    n = x.size
    n_harm = 10                     # number of harmonics in model
    t = np.arange(0, n)
    p = np.polyfit(t, x, 1)         # find linear trend in x
    x_notrend = x - p[0] * t        # detrended x
    x_freqdom = fft.fft(x_notrend)  # detrended x in frequency domain
    f = fft.fftfreq(n)              # frequencies
    
    # В этой строчке добавляем list()
    # indexes = range(n)
    indexes = list(range(n))
    
    # sort indexes by frequency, lower -> higher
    indexes.sort(key = lambda i: np.absolute(f[i]))
 
    t = np.arange(0, n + n_predict)
    restored_sig = np.zeros(t.size)
    for i in indexes[:1 + n_harm * 2]:
        ampli = np.absolute(x_freqdom[i]) / n   # amplitude
        phase = np.angle(x_freqdom[i])          # phase
        restored_sig += ampli * np.cos(2 * np.pi * f[i] * t + phase)
    return restored_sig + p[0] * t
```

```python
n_pred = 365
pred = fourierExtrapolation(df['t'], n_pred)[-n_pred:]

t = list(range(len(df['t'])))
t_pred = list(range(len(t), len(t) + n_pred))

plt.plot(t, df['t'].values)
plt.plot(t_pred, pred)
```

Если взять другое количество n_harm, то получится более зашумленая кривая.
